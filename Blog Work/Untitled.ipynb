```python
from wordcloud import WordCloud
```

```python
import matplotlib.pyplot as plt
% matplotlib inline
```

```python
text = ['Python ', 'Python ', 'Python ', 'Python ', 'Python ', 'Algorithm ', ' Computation ', 'Distributed Programming ', 'Data Science ',
        'Data Data ','Data Data ','Data Data ',
       'Data Science ', 'Decision Science ', 'Decision Science ','Decision Science ', 
        'Environmental Science ', 'Environmental Data ','Environmental Science ', 'Behavioral science ', 'Behavioral science ',
       'Behavioral science ', 'Nudging ', 'Nudges ', 'Nudge ','Photography ']
```

```python
''.join(text)

```

```python
wordcloud = WordCloud(background_color='white', width=1200, height=400).generate(''.join(text))
plt.figure()
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.show()
```
